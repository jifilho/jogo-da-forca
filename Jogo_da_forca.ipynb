```python
import getpass

def jogar_forca():
  partesBoneco = [
      '''
      ------
      |    |
      |
      |
      |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |
      |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |    |
      |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |   -|
      |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |   -|-
      |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |   -|-
      |   |
      -
      ''',
      '''
      ------
      |    |
      |    O
      |   -|-
      |   | |
      -
      '''
  ]


  palavraForca = ""
  linhaForca = []
  tentativasRestantes = 6
  bonecoRestante = 0

  print("*--------------------------------------------------------------------------------------------------------------------------*")
  print(">                                          BEM-VINDO AO JOGO DA FORCA!                                                     <")
  print(">                                     Este jogo deve ser jogado em dupla, viu?                                             <")
  print("> As regras do jogo são: Você irá digitar o nome da palavra e seu amigo irá tentar adivinhar qual palavra que é. Vamos lá? <")
  print("*--------------------------------------------------------------------------------------------------------------------------*")

  while True:
      escolhaPalavra = getpass.getpass("> Escreva uma palavra a ser adivinhada: ")

      if len(escolhaPalavra) >= 3:
          palavraForca = escolhaPalavra
          print("Perfeito. Vamos começar!")
          break
      else:
          print("A palavra deve ter no mínimo 3 caracteres. Tente novamente.")

  linhaForca = ["_"] * len(palavraForca)

  while "_" in linhaForca:
      print(partesBoneco[bonecoRestante])
      #print("Tentativas restantes: ", tentativasRestantes)
      print(' '.join(linhaForca))
      letraJogador = input("> Digite uma letra: ")

      acertou = False

      for i, letra in enumerate(palavraForca):
          if len(letraJogador) != 1:
              print("> Valor inválido. Digite apenas uma letra.")
              break

          if letraJogador == linhaForca[i]:
              print("Você já digitou essa palavra!")
              break

          if letraJogador == letra:
              linhaForca[i] = letra
              acertou = True

      if not acertou and len(letraJogador) == 1:
          tentativasRestantes = tentativasRestantes - 1
          bonecoRestante = bonecoRestante + 1

          if tentativasRestantes > 0:
              print("> Letra não encontrada. Tente novamente.")
          else:
              print(partesBoneco[bonecoRestante])
              print("*--------------------------------------------------------------------------------------------------------------------------*")
              print(">                                                 Que pena, você perdeu!                                                   <")
              print("*--------------------------------------------------------------------------------------------------------------------------*")
              break

  if tentativasRestantes > 0:
    print(partesBoneco[bonecoRestante])
    print("*--------------------------------------------------------------------------------------------------------------------------*")
    print(">                                        Parabéns! Você adivinhou a palavra:", palavraForca,"                                          <")
    print("*--------------------------------------------------------------------------------------------------------------------------*")

while True:
    jogar_forca()
    jogar_novamente = input("Deseja jogar novamente? (Digite 's' para sim, 'n' para não): ")
    if jogar_novamente.lower() != 's':
      print("Obrigado por jogar!")
      break


#ATUALIZAÇÕES 24/01:
#- Criado a validação quando alguém digitar mais de um caractere para o campo letraJogador.
#- Reduzido redundâncias no laço while, corrigido erro que retornava "Letra não encontrada. Tente novamente" mesmo com 0 tentativas restantes.
#- Criado uma interface gráfica simples em ASCII para ficar visualmente mais bonito.
#= Criado interface gráfica em ASCII do boneco do jogo, substituindo o contador de tentativas restantes pela representação visual
#- Criptografado a palavra na variável escolhaPalavra utilizando a biblioteca getpass
#- Inserido verificação para a palavra possuir, no mínimo, 3 letras

#ATUALIZAÇÕES 26/01:
#- A lista encriptada das palavras agora aparecem no início do jogo
#- Criado validação para apresentar erro quando um jogador repetir a letra

#ATUALIZAÇÕES 29/01:
#- Criado função para reiniciar o jogo. Coloquei o código "principal" dentro de uma função e fiz um while no fim
```
